**The List in Python**  
Lists are used to store multiple items in a single variable.

Lists are one of 4 built-in data types in Python used to store collections of data, the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.

Lists are created using square brackets:

```python
#let's create the list and check and do verious list methods
'''
properties of the List -->
->Ordered
->Changeable
->Allow Duplicates
->Can access with positive and the negative indexing both
->Data-Type is: <class 'list'>
'''
listOfFruit = ["apple","banana","cherry","mango","orange","guava"]

#Accessing the list items
indexZero = listOfFruit[0]
indexOne = listOfFruit[1]
indexLast = listOfFruit[len(listOfFruit)-1]
print("First index: ",indexZero,", Second index: ",indexOne,", last index: ",indexLast)
print("the length of the list is: ",len(listOfFruit))
print()

#Negative indexing 
'''
index: -1, last
index: (len(list)*(-1)), tell's zeroth index
index: (len(list)*(-1)+1) , tell's the first index.
.........so on.
'''
indexZero = listOfFruit[(len(listOfFruit)*(-1))]
indexOne = listOfFruit[(len(listOfFruit)*(-1))+1]
indexLast = listOfFruit[-1]
print("First index: ",indexZero,", Second index: ",indexOne,", last index: ",indexLast)
print("the length of the list is: ",len(listOfFruit))
print()

#Range of indexess
rangeOFList = listOfFruit[1:4]
print("From index one to index foure elements aer: ",rangeOFList)
rangeOFList = listOfFruit[:3]  #if first is argument is not given it consider it is zero'th index 
print(rangeOFList)
rangeOFList = listOfFruit[3:]
print(rangeOFList)
print()

#Range of negative indexess
rangeOFList = listOfFruit[-4:-1]
print("From index one to index foure elements aer: ",rangeOFList)
rangeOFList = listOfFruit[:-3]  #if first is argument is not given it consider it is zero'th index 
print(rangeOFList)
rangeOFList = listOfFruit[-3:]
print(rangeOFList)
print()

#Changing the list's values
   #--> with the normal indexing
listOfFruit[0] = "Lucky"
listOfFruit[len(listOfFruit)-1] = "febvriot"
print(listOfFruit)
   #--> with the negative indexing
listOfFruit[(len(listOfFruit)*(-1))] = "apple"
listOfFruit[-1 ] = "guava"
print(listOfFruit)
print()

#Insert the value in the list (the insert method only inset the item in the specify location of the list)
       # --> syntex: listNmae.insert(index,value)
listOfFruit.insert(len(listOfFruit),"kivi")
listOfFruit.insert((len(listOfFruit)*(-1)),"water-mellon")
print(listOfFruit)
print()

#Append method
listOfFruit.append("coconut")
print(listOfFruit)
print()

#Extened the list (append the anothe list in the existing list)
         #syntex: TheListWhereYouWantTOAppend.extend(TheListYouAdd)
secondListOfFruits = ["stravery","grapes","balck-berry","papaya"]
listOfFruit.extend(secondListOfFruits)
print(listOfFruit)
print()

#Remove the specify item with the name
     #syntex: listName.remove(itemName)
listOfFruit.remove("kivi")
print(listOfFruit)
print()

#Remove at the specify index
listOfFruit.pop(len(listOfFruit)-1)
print(listOfFruit)
print()

#Remove the element with the help of the 'del' keyword
     #syntex: del listName[index]
print(secondListOfFruits)
del secondListOfFruits[len(secondListOfFruits)-1]
print(secondListOfFruits)
print()    

#Delet all the elements in the list
secondListOfFruits.clear()
print(secondListOfFruits)  #-->now it become null list
print()

#Sorting the lis
      #Syntex: listName.sort() --> sorts alphabetically by default assending order
      #        listName.sort(reverce = True) --> for the decending order
listOfFruit.sort()
print("List sorted assecnding order: ",listOfFruit)
listOfFruit.sort(reverse=True)
print("Sorted as decending order: ",listOfFruit)
print()

thirdList = listOfFruit[:int(len(listOfFruit)/2)]

#To the reverce the list
print("The thirdList: ",thirdList)
thirdList.reverse()   #not sort the alphabetically direct reverce as 1st gose -->last index and last goes -->1st index
print("The reverce of the thirdlist: ",thirdList)
print()

#To copy the list in the other variable 
'''
we can not copy list as
      list1 = list2  --> not copy list2 in the list1 it gives us the an error
      so using copy() we can do this
    '''
fourthList = thirdList.copy()
print("thirdList is: ",thirdList," and \nfoutrth lis is which is contentes all items as in the list3: ",fourthList)
''' 
anothe way to copy any list is
   use the built in method list()
    listName_where_you_want_toCopy =list(ListWhichCopy)
'''
listFifth = list(fourthList)
print("fifth list is: ",listFifth)
print()

#join method for joining the two list togather
'''
the extend() method also adds the two lists togather.
-->list1.extend(list2)
But the extent method is appends the list2 into list1
  if you want to add list2 into list1 but store into anothr list then use the '+' operator
'''
listSixth = listFifth+fourthList
print("listSixth is: ",listSixth)
```

A tuple is a collection which is ordered and unchangeable.

**Set in the Pytho**  
 Sets are used to store multiple items in a single variable.

Set is one of 4 built-in data types in Python used to store collections of data, the other 3 are List, Tuple, and Dictionary, all with different qualities and usage.

A set is a collection which is unordered, unchangeable*, and unindexed.

* Note: Set items are unchangeable, but you can remove items and add new items.

Sets are written with curly bracketss


```python
'''
Lets creates the list in the python
using the '{ }' brackets we can create the set in python

    ## PROPERTIES OF THE SET ##
    -->Unordered (means every-time you see diffrent diffrent location any set values)
    -->Unchangeable (but can add and removes the items)
    -->Duplicates are not allows
    -->Can not access using any indexing (neg. or posi. indexing)
    -->Data-type is <class 'set'>
                Let's create the set: 
'''
mySet = {100,0.22,'python',True,0,-3878}
print("The set elements is: ",mySet)
print("one more time printing the same set: ",mySet)
print()

#Adding the elements in the set
mySet.add(5903)
mySet.add(100)  #This will not add again becouse the 100 is already present in the set
print("After adding the set is: ",mySet)
print()

#Adding the set in the exisitng set --> it can done using the update() method
mySet2 = {400,'a',"alphabets"}
mySet.update(mySet2)
print("After using the update method in the set is: ",mySet)
print()

#Removing the items in the set
        # we can use the veriouse method to remove the elements like this
#using the value
mySet.remove(100)
print(mySet)
#using discart method --> if giving value is not in the set then the removes will give en error but discart() will not give an error
mySet.discard(549833)
print(mySet)
#pop()  --> this will removes the random element 
mySet.pop()
print("After the poping the element: ",mySet)
#clear() method
mySet2.clear()
print("After appliying the clear() in the mylist2: ",mySet2)
print()

#Set Properties in the meths
''' 
we can also use the set properties as:
1) Intersection
2) Union
    ect.
'''
set1 = {1,2,3,'a','b','c'}
set2 = {4,2,5,'a','d','c'}
''' 
->use the union methd if you want both sets element in the new set
->use update method if you want to append the set in the another set
'''
set3 = set1.union(set2)
print("The set3 is: ",set3)
set1.update(set2)
print("After the using of update methods: ",set1)
print()

''' 
if you want to take the only comman elements in the both set use intersection methods
1) if you dont want the new set use-->  intersection_update() method
2) if you want to the new set use--> insersection() methods

if you want to keep all the values but not want to get the same values in both the sets
1) symmetric_diffrence_update() --> for the not the net set creation
2) symmetric_diffrence() --> for new set creation

'''
set1 = {1,2,3,'a','b','c'}
set2 = {4,2,5,'a','d','c'}
set4 = set1.intersection(set2)
print("Values in the set4 after applying the intersection method: ",set4)
set1.intersection_update(set2)
print("The set1 after appling the intersection_update() method: ",set1)
set1 = {1,2,3,'a','b','c'}
set2 = {4,2,5,'a','d','c'}
set4 = set1.symmetric_difference(set2)
print("After useing the symmetric_diffrence() method: ",set4)
set1.symmetric_difference_update(set2)
print("After appliying the symmetric_diffrence_update() method set1 is: ",set1)
```

Dictionaries are used to store data values in key:value pairs. 
A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

Dictionaries are written with curly brackets, and have keys and values:
